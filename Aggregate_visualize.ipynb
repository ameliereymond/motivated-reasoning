```python
import pandas as pd
import numpy as np
import re
```

# Aggregate DataFrames

```python
def process_answers(llm_df):
    #mist_replicated = pd.concat([mist]*100, ignore_index = True)
    df = llm_df.dropna()
    # print(llm_df)
    df['perceived_accuracy'] = df['perceived_accuracy'].apply(lambda x: int(match.group()) if (match := re.search(r'\b\d+\b', x)) else None)
    df['confidence'] = df['confidence'].apply(lambda x: int(match.group()) if (match := re.search(r'\b\d+\b', x)) else None)
    df.dropna(subset = ['perceived_accuracy', 'confidence'], inplace = True)
    df[['perceived_accuracy', 'confidence']] = df[['perceived_accuracy', 'confidence']].astype(int)

    return df
```

# COT 

## Political

### Llama 3

```python
llama_df_r = pd.read_csv('data/llama3.1_republican_cot_MIST.csv')
llama_df_d = pd.read_csv('data/llama3.1_democrat_cot_MIST.csv')
llama_df_r = process_answers(llama_df_r)
llama_df_d = process_answers(llama_df_d)


llama_df = pd.concat([llama_df_r, llama_df_d], ignore_index = True)
llama_df['political_persona'] = ['Republican']*llama_df_r.shape[0] + ['Democrat']*llama_df_d.shape[0]
llama_df.to_csv('data/llama3.1_republican_democrat.csv')
```

### Llama 2
